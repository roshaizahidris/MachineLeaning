```python
# Supervised ML
# Classification-->Logictic Regression

# Sigmoid Function
import numpy as np
from sklearn.linear_model import LogisticRegression

# step 2 : create the dataset
X = np.array([[1],[2],[3],[4],[5],[6]])   # hours_studied
y = np.array([0,0,0,1,1,1])               # passed or not


# step 3: create and train the model
model=LogisticRegression()
model.fit(X,y)
```

```python
# predict_proba() stands for predict probabilities
y_prob=model.predict_proba(X)[:,1]   #aqll tows from column 1
y_prob
```

```python
# get the slope and intercept
a=model.coef_[0][0]         # slope  coef_ is a array in  scikit-learn
b=model.intercept_[0]       # intercept

print(f'Slope :{a}')
print(f'Intercept : {b}')
```

```python
# Logistic Regression- Predicting Diabetes

import pandas as pd
data=pd.read_csv('diabetes.csv')
data.head()
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
x=data.drop('Outcome',axis=1)   #axis=1 means column, axis=0 is row
x
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
# Select independent and Dependent variable
x=data.drop('Outcome',axis=1)
y=data['Outcome']
```

```python
# train_test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Model Development
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
# predict_proba() stands for predict probabilities
y_prob=model.predict_proba(x_test)[:,1]  # all rows from column 1
y_prob
```

```python
# get the predicted values
y_pred = model.predict(x_test)

# show the predictions
y_pred
```

```python
ev1=pd.DataFrame({'Probabilities': y_prob,'Predicted': y_pred})
ev1.head(20)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)

#  Total Observation : 154
#                  Actual_Pos    Actual_Neg
# Predicted_Pos       TP=78        FP=21      78+37==>115[Good Prediction]
# Predicted_Neg       FN=18        TN=37      21+18==>39[Wrong Prediction]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(cm,annot=True,cmap='Blues')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

```python
print('Classification Report\n')
print(classification_report(y_test,y_pred))
```
