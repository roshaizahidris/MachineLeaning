```python
sns.set_theme()
df=sns.load_dataset('titanic')
print('Seaborn:',sns.__version__)
print('Shape:',df.shape)
df.head
```

```python
#countplot()
plt.figure(figsize=(6,4))
colors=['red','green']
sns.countplot(data=df,x='survived',palette=colors)
plt.title('Numbers of Survival (0=no,1=yes)')
plt.show()
```

```python
# to avoid any problem in future we modified the coding follow what the warning instruc.
plt.figure(figsize=(6,4))
colors=['red','green']
sns.countplot(data=df,x='survived',hue='survived',palette=colors)
plt.title('Numbers of Survival (0=no,1=yes)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
colors=['red','green']
sns.countplot(data=df,x='survived',hue='sex',palette=colors)
plt.title('Numbers of Survival (0=no,1=yes)')
plt.show()
```
